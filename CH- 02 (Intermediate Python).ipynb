## Global Keyword

`````python
x = 24
def increase():
    x  = x + 1

increase()
`````

** why?
- Since `x = ` is initialized and it treats `x` as local variable, 
- But doesnot find it's value, so the operation `x+1` fails. 
- the local variable has not been assigned yet.

* we add `global ` to make `x` treated as global and not `local` 

`````python
x = 24
def increase():
    global x
    x  = x + 1
    print(x)

increase()
`````

** Guess the output

`````python
x = 69

def outer():

    x = 96

    def inner():

        global x

        print(x)

    inner()

outer()

`````

* `global` refer to module-level global, not enclosing function's variable. 

When Should You Use `global`?
- But we avaoid, 
- Debugging becomes difficult. 
- Accidental modification. 

`````python
x = 1

def outer():

    x = 2

    def inner():

        global x

        x = 3

    inner()

    print(x)

outer()

print(x)
`````

* Because, inside outer function, `x = 2` is the (L) local , so it prints `2`. 
* But for outside, there is no local, it then looks for `global` which was set to `2` and modified to `3`. 

## `Non-Local` keyword

`````python
def outer():
    x = 10
    def inner():
        x = x + 1
        print(x)
    inner()
outer()
`````

`UnboundLocalError`: cannot access local variable 'x' where it is not associated with a value
- Because, `x =` is initiated as `local` variable, and it tries to access its value. 


### To Solve this issue???

We need to say python not to create this `local variable`.  
- We have `nonlocal`

`````python
def outer():
    x = 10
    def inner():
        nonlocal x
        x = x + 1
        print(x)
    inner()
outer()
`````

###  What does `nonlocal` do??
- It tells python interpreter that this variable, belongs to the nearest enclosing function. 

`````python
x = 1
def outer():
    x = 2

    def inner():
        nonlocal x
        x = 3

    inner()
    print(x)

outer()
print(x)

`````

`3` because, the enclosed `2` is modified by accessing it as `local`.  
But `nonlocal` cant modify global `x = 1`, 

`````python
x = 10
def outer():
    nonlocal x
    
`````

Because there is no `enclosing function containing `x`. 

### Why `nonlocal` needed?
- Otherwise, enclosing functions variables can't be modified.
